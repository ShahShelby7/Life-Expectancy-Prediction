### Area plot

```python
import matplotlib.pyplot as plt

# Sorting the data by 'percentage expenditure'
sorted_data = df.sort_values(by='percentage expenditure')

# Calculating the cumulative sum of 'percentage expenditure'
cumulative_sum = sorted_data['percentage expenditure'].cumsum()

# Plotting a basic area chart
plt.figure(figsize=(8, 6))
plt.fill_between(sorted_data.index, cumulative_sum, color='skyblue', alpha=0.4)
plt.plot(sorted_data.index, cumulative_sum, color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Cumulative Sum of percentage expenditure ')
plt.xlabel('Index')
plt.ylabel('Cumulative Sum')
plt.grid(True)
plt.show()
```

### 1.LINEAR REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
x=df.drop(["Life expectancy ","Country","Status","Year"],axis=1)
y=df["Life expectancy "]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=70)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
reg.fit(x_train,y_train)
```

```python
pred=reg.predict(x_test)
pred
```

```python
y_pred = reg.predict(x_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:",r2)
```

### 2.Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print("Root Mean Squared Error:", rmse)
print("R-squared:",r2)
```
